### Step 2: Calculate TPR and FPR at Each Threshold

Using the sorted values, calculate the True Positives (TP), False Positives (FP), True Negatives (TN), and False Negatives (FN) for each threshold and then compute TPR and FPR.

1. **Threshold: 0.8**
   - TP = 1, FP = 0, FN = 3, TN = 3
   - TPR = 1/4 = 0.25
   - FPR = 0/3 = 0

2. **Threshold: 0.6**
   - TP = 2, FP = 1, FN = 2, TN = 2
   - TPR = 2/4 = 0.5
   - FPR = 1/3 = 0.33

3. **Threshold: 0.5**
   - TP = 3, FP = 1, FN = 1, TN = 2
   - TPR = 3/4 = 0.75
   - FPR = 1/3 = 0.33

4. **Threshold: 0.3**
   - TP = 4, FP = 1, FN = 0, TN = 2
   - TPR = 4/4 = 1.0
   - FPR = 1/3 = 0.33

5. **Threshold: 0.2**
   - TP = 4, FP = 2, FN = 0, TN = 1
   - TPR = 4/4 = 1.0
   - FPR = 2/3 = 0.67

6. **Threshold: 0.1**
   - TP = 4, FP = 3, FN = 0, TN = 0
   - TPR = 4/4 = 1.0
   - FPR = 3/3 = 1.0

Using pairs of (FPR,TPR)coordinates:


(0,0.25),(0.33,0.5),(0.33,0.75),(0.33,1.0),(0.67,1.0),(1.0,1.0)

Interval 1: (0,0.33)

Interval 2: (0.33,0.33)

Interval 3:  (0.33,0.33)

Interval 4:  (0.33,0.67)

Interval 5:  (0.67,1.0)

For each interval, calculaing the area of the trapezoid:

Area= (FPR2 − FPR1) × (TPR1 + TPR2)/ 2


Interval 1: Area = 0.12375

Interval 2: Area= 0(no width, so area is zero)

Interval 3: Area = 0

Interval 4: Area = 0.34

Interval 5: Area = 0.33


AUC=0.12375+0+0+0.34+0.33=0.79375

```python
import numpy as np

def calculate_auc(labels, predictions):
    # Combine labels and predictions into a list of tuples
    data = list(zip(predictions, labels))

    # Sort the data by predictions in descending order
    data.sort(key=lambda x: x[0], reverse=True)

    # Separate sorted predictions and labels
    sorted_preds, sorted_labels = zip(*data)

    # Calculate total positive and negative samples
    total_pos = sum(sorted_labels)
    total_neg = len(sorted_labels) - total_pos

    # Initialize variables
    false_pos = 0
    true_pos = 0
    prev_false_pos = 0
    prev_true_pos = 0
    auc = 0

    # Calculate AUC
    for i in range(len(sorted_labels)):
        if i > 0 and sorted_preds[i] != sorted_preds[i-1]:
            # Calculate trapezoid area
            auc += (false_pos - prev_false_pos) * (true_pos + prev_true_pos) / 2
            prev_false_pos = false_pos
            prev_true_pos = true_pos

        if sorted_labels[i] == 1:
            true_pos += 1
        else:
            false_pos += 1

    # Add final trapezoid
    auc += (false_pos - prev_false_pos) * (true_pos + prev_true_pos) / 2

    # Normalize AUC
    auc /= (total_pos * total_neg)

    return auc

# Example usage
labels = [1, 0, 1, 1, 1, 0, 0]
predictions = [0.8, 0.6, 0.6, 0.5, 0.3, 0.2, 0.1]

auc = calculate_auc(labels, predictions)
print(f"AUC: {auc:.4f}")
```

## Task 3 [2 points]

Your user got such movies as recommendation with such positions:

| Position                     | Movies                               |
|-----------------------------|-------------------------------------------|
| 1                      | Movie A |
| 2             | Movie C   |
| 3                      | Movie F  |
| 4                   | Movie D |
| 6                   | Movie E |

The user gave such scores for these movies:

| Movie                     | Score                               |
|-----------------------------|-------------------------------------------|
| Movie A                      | 3 |
| Movie C           | 2  |
| Movie F                      | 0  |
| Movie D                   | 0 |
| Movie E                  | 1 |

Calculate NDCG.

DCG= i=1∑ (rel)i/ (log 2 (i+1))

Scores: Movie A = 3, Movie C = 2, Movie F = 0, Movie D = 0, Movie E = 1

Positions: Movie A (1), Movie C (2), Movie F (3), Movie D (4), Movie E (6)

Calculating  DCG for Each Position

Movie A:  3/ log 2 (1+1) = 3/1 = 3

Movie C: 2/ log 2 (2+1) ≈ 1.2619

Movie F: 0/ log 2 (3+1) = 0

Movie D:  0/ log 2 (4+1) = 0

Movie E:  1/ log 2 (6+1) ≈ 0.3562

DCG=3+1.2619+0+0+0.3562=4.6181


**Ideal DCG**

First position (score = 3):  3/ log 2 (1+1) = 3/1 = 3

Second position (score = 2): 2/ log 2 (2+1) ≈ 1.2619

Third position (score = 1): 1/ log 2 (3+1) = 0-5

Fourth position (score = 0):  0/ log 2 (4+1) = 0

Fifth position (score = 0):  0/ log 2 (5+1) ≈ 0.

IDCG=3+1.2619+0.5+0+0=4.7619


**NDCG** =  IDCG / DCG = 4.7619 / 4.6181 ≈0.97


```python
import math

def dcg_at_k(relevances, k):
    """Calculate Discounted Cumulative Gain at k"""
    dcg = 0
    for i, rel in enumerate(relevances[:k], 1):
        dcg += (2**rel - 1) / math.log2(i + 1)
    return dcg

def ndcg_at_k(relevances, k):
    """Calculate Normalized Discounted Cumulative Gain at k"""
    dcg = dcg_at_k(relevances, k)
    ideal_relevances = sorted(relevances, reverse=True)
    idcg = dcg_at_k(ideal_relevances, k)
    return dcg / idcg if idcg > 0 else 0

# Given data
recommendations = [
    (1, "Movie A"),
    (2, "Movie C"),
    (3, "Movie F"),
    (4, "Movie D"),
    (6, "Movie E")
]

scores = {
    "Movie A": 3,
    "Movie C": 2,
    "Movie F": 0,
    "Movie D": 0,
    "Movie E": 1
}

# Prepare relevance scores in the order of recommendations
relevances = [scores[movie] for _, movie in recommendations]

# Calculate NDCG
k = len(recommendations)  # Consider all recommended items
ndcg = ndcg_at_k(relevances, k)

print(f"Relevance scores: {relevances}")
print(f"NDCG@{k}: {ndcg:.4f}")

# Calculate NDCG@3 and NDCG@5 for comparison
print(f"NDCG@3: {ndcg_at_k(relevances, 3):.4f}")
print(f"NDCG@5: {ndcg_at_k(relevances, 5):.4f}")
```

```python
import matplotlib.pyplot as plt
from lightfm import LightFM
from lightfm.data import Dataset
from lightfm.cross_validation import random_train_test_split


# Filter implicit positive ratings
implicit_ratings = ratings[ratings['rating'] >= 4]

# Initialize dataset
dataset = Dataset()
dataset.fit(implicit_ratings['user_id'].unique(), implicit_ratings['movie_id'].unique())

# Build user-item interactions matrix
(interactions, weights) = dataset.build_interactions(implicit_ratings[['user_id', 'movie_id']].values)

# Build user and item features
user_features = dataset.build_user_features([(x, []) for x in implicit_ratings['user_id'].unique()])
item_features = dataset.build_item_features([(x, []) for x in implicit_ratings['movie_id'].unique()])


# Split data into train and test (80/20 split)
train_interactions, test_interactions = random_train_test_split(interactions, test_percentage=0.2)

# Model parameters
epochs = 30
threads = 4
auc_scores_bpr, auc_scores_warp = [], []

# Train with BPR loss
model_bpr = LightFM(loss='bpr')
for epoch in range(epochs):
    model_bpr.fit_partial(train_interactions, user_features=user_features, item_features=item_features, num_threads=threads)
    auc_train = auc_score(model_bpr, train_interactions, user_features=user_features, item_features=item_features, num_threads=threads).mean()
    auc_test = auc_score(model_bpr, test_interactions, user_features=user_features, item_features=item_features, num_threads=threads).mean()
    auc_scores_bpr.append((auc_train, auc_test))
    print(f"BPR - Epoch {epoch+1}/{epochs}: Train AUC={auc_train:.3f}, Test AUC={auc_test:.3f}")

# Train with WARP loss
model_warp = LightFM(loss='warp')
for epoch in range(epochs):
    model_warp.fit_partial(train_interactions, user_features=user_features, item_features=item_features, num_threads=threads)
    auc_train = auc_score(model_warp, train_interactions, user_features=user_features, item_features=item_features, num_threads=threads).mean()
    auc_test = auc_score(model_warp, test_interactions, user_features=user_features, item_features=item_features, num_threads=threads).mean()
    auc_scores_warp.append((auc_train, auc_test))
    print(f"WARP - Epoch {epoch+1}/{epochs}: Train AUC={auc_train:.3f}, Test AUC={auc_test:.3f}")


# Plotting AUC scores over epochs
bpr_train_scores, bpr_test_scores = zip(*auc_scores_bpr)
warp_train_scores, warp_test_scores = zip(*auc_scores_warp)

plt.figure(figsize=(12, 6))
plt.plot(range(1, epochs + 1), bpr_train_scores, label="BPR Train AUC", linestyle='-', marker='o', color='blue')
plt.plot(range(1, epochs + 1), bpr_test_scores, label="BPR Test AUC", linestyle='--', marker='o', color='cyan')
plt.plot(range(1, epochs + 1), warp_train_scores, label="WARP Train AUC", linestyle='-', marker='x', color='red')
plt.plot(range(1, epochs + 1), warp_test_scores, label="WARP Test AUC", linestyle='--', marker='x', color='orange')

plt.xlabel('Epochs')
plt.ylabel('AUC Score')
plt.title('AUC Score Progression for BPR and WARP Losses')
plt.legend()
plt.grid()
plt.show()
```

```python
import gzip
import json
import pandas as pd
from implicit.als import AlternatingLeastSquares
from implicit.bpr import BayesianPersonalizedRanking
from implicit.evaluation import train_test_split, precision_at_k, mean_average_precision_at_k
from scipy.sparse import coo_matrix
import numpy as np

# Load the JSONL file and extract user_id, item_id, and rating
file_path = '/content/drive/MyDrive/All_Beauty.jsonl'

# Parse JSONL to create a DataFrame
data = []
with open(file_path, 'r') as fp:
    for line in fp:
        record = json.loads(line.strip())
        # Check if required keys are present
        if 'user_id' in record and 'asin' in record and 'rating' in record:
            data.append((record['user_id'], record['asin'], record['rating']))

df = pd.DataFrame(data, columns=['user_id', 'item_id', 'rating'])

# Create user and item mappings
user_ids = df['user_id'].astype('category')
item_ids = df['item_id'].astype('category')

user_map = dict(enumerate(user_ids.cat.categories))
item_map = dict(enumerate(item_ids.cat.categories))

df['user'] = user_ids.cat.codes
df['item'] = item_ids.cat.codes

# Create the sparse matrix
sparse_matrix = coo_matrix((df['rating'], (df['user'], df['item'])))

# Train-test split using 80% for training and 20% for testing
train, test = train_test_split(sparse_matrix, train_percentage=0.8, random_state=42)


#function to evaluate models
def evaluate_model(model, train, test):
    model.fit(train)
    p_at_k = precision_at_k(model, train, test, K=10)
    map_at_k = mean_average_precision_at_k(model, train, test, K=10)
    return np.mean(p_at_k), map_at_k

#train and evaluate als model
als_results = []

for factors in [50, 100, 200]:
    for regularization in [0.01, 0.1, 1.0]:
        for iterations in [10, 20, 50]:
            als = AlternatingLeastSquares(factors=factors, regularization=regularization, iterations=iterations)
            precision, map_score = evaluate_model(als, train, test)
            als_results.append({
                'model': 'ALS',
                'factors': factors,
                'regularization': regularization,
                'iterations': iterations,
                'precision@10': precision,
                'MAP@10': map_score
            })

als_df = pd.DataFrame(als_results)
print(als_df.sort_values('MAP@10', ascending=False).head())

#train and evaluate bpr model
bpr_results = []

for factors in [50, 100, 200]:
    for learning_rate in [0.01, 0.05, 0.1]:
        for iterations in [10, 20, 50]:
            bpr = BayesianPersonalizedRanking(factors=factors, learning_rate=learning_rate, iterations=iterations)
            precision, map_score = evaluate_model(bpr, train, test)
            bpr_results.append({
                'model': 'BPR',
                'factors': factors,
                'learning_rate': learning_rate,
                'iterations': iterations,
                'precision@10': precision,
                'MAP@10': map_score
            })

bpr_df = pd.DataFrame(bpr_results)
print(bpr_df.sort_values('MAP@10', ascending=False).head())

#select best hyperparameter
best_als = als_df.loc[als_df['MAP@10'].idxmax()]
best_bpr = bpr_df.loc[bpr_df['MAP@10'].idxmax()]

print("Best ALS hyperparameters:")
print(best_als)
print("\nBest BPR hyperparameters:")
print(best_bpr)

#train final models with best hyperparameter
final_als = AlternatingLeastSquares(
    factors=int(best_als['factors']),
    regularization=best_als['regularization'],
    iterations=int(best_als['iterations'])
)

final_bpr = BayesianPersonalizedRanking(
    factors=int(best_bpr['factors']),
    learning_rate=best_bpr['learning_rate'],
    iterations=int(best_bpr['iterations'])
)

final_als.fit(sparse_matrix)
final_bpr.fit(sparse_matrix)
```
